```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create sample accident dataset
np.random.seed(42)

data = {
    "Accident_ID": range(1, 501),
    "Location": np.random.choice(["City Center", "Highway", "Suburb", "Rural"], 500),
    "Weather": np.random.choice(["Clear", "Rain", "Fog", "Snow"], 500),
    "Road_Condition": np.random.choice(["Dry", "Wet", "Icy"], 500),
    "Time_of_Day": np.random.choice(["Morning", "Afternoon", "Evening", "Night"], 500),
    "Severity": np.random.choice(["Minor", "Major", "Fatal"], 500),
    "Latitude": np.random.uniform(12.9, 13.1, 500),
    "Longitude": np.random.uniform(77.5, 77.7, 500)
}

df = pd.DataFrame(data)

print("Sample Accident Dataset")
print(df.head())

# Dataset information
print("\nDataset Info:")
print(df.info())

# Accidents by Weather
plt.figure(figsize=(6,4))
sns.countplot(x="Weather", data=df)
plt.title("Accidents by Weather Condition")
plt.show()

# Accidents by Road Condition
plt.figure(figsize=(6,4))
sns.countplot(x="Road_Condition", data=df)
plt.title("Accidents by Road Condition")
plt.show()

# Accidents by Time of Day
plt.figure(figsize=(6,4))
sns.countplot(x="Time_of_Day", data=df)
plt.title("Accidents by Time of Day")
plt.show()

# Severity vs Weather
plt.figure(figsize=(6,4))
sns.countplot(x="Weather", hue="Severity", data=df)
plt.title("Accident Severity vs Weather")
plt.show()

# Hotspot Visualization (Latitudeâ€“Longitude)
plt.figure(figsize=(6,6))
plt.scatter(df["Longitude"], df["Latitude"], alpha=0.5)
plt.title("Accident Hotspots")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()

# Contributing Factors Heatmap
pivot_table = pd.pivot_table(
    df,
    values="Accident_ID",
    index="Weather",
    columns="Road_Condition",
    aggfunc="count"
)

plt.figure(figsize=(6,4))
sns.heatmap(pivot_table, annot=True, fmt="g", cmap="Blues")
plt.title("Accidents by Weather and Road Condition")
plt.show()

print("\nTASK 04 COMPLETED SUCCESSFULLY!")
```
